## Do opioid overdoes tend to be associated with less affluent areasâ€”that is, areas wherefamilies have lower incomes?

```python
"""Format of the scatterplot method"""
fig, ax = plt.subplots()
fig.suptitle('Opioid Overdoes vs. Median Household Income', fontsize = 14) #Graph Title
fig.set_size_inches(7,5)

ax.scatter(df_opioids['MedianHouseholdIncome'],df_opioids['FPOO-Rate'], alpha = 0.5, c = 'b' ) #alpha controls dot transparency: 1 = solid , 0 =no mark, b stand for blue
ax.xaxis.set_label_text('Median Household Income') # x axis caption
ax.yaxis.set_label_text('Opioids Overdoes Rate - per 100,000 County Residents') #y axis caption
plt.show()
```

In order to answer this question, I needed to visualize the data by using a scatterplot. The Opioid scatterplot for Opioids Overdoes versus Median Household Income shows tendency higher opioids overdoes in less affluent areas. 

## Some people who start with opioid addictions are reported to transition to heroin use.What is the relationship in Virginia counties between opioid overdoses and heroinoverdoses? 

```python
"""Format of the scatterplot method"""
fig, ax = plt.subplots()
fig.suptitle('Do Opioid Overdoes influence Heroine Overdoes Rate in Virginia',fontsize = 14) #Graph Title
fig.set_size_inches(7,5)

ax.scatter(df_opioids['FPOO-Rate'],df_opioids['FFHO-Rate'], alpha = 0.5, c = 'g' ) #alpha controls dot transparency: 1 = solid , 0 =no mark, b stand for blue
ax.xaxis.set_label_text('Opioids Overdoes Rate - per 100,000 County Residents') # x axis caption
ax.yaxis.set_label_text('Heroine Overdoes Rate - per 100,000 County Residents') #y axis caption
plt.show()
```

The data on the scatterplot show veary weak to no correlation between Opioids and Heroine 

```python
"""Format of the bar Chart method"""
fig, ax = plt.subplots()
fig.suptitle('Opioid Overdoes vs. Heroine Overdoes by County', fontsize = 24) #Graph Title
fig.set_size_inches(20,10)
index = np.arange(len(df_opioids['CountyName']))
bar_width = 0.5

Opioids = plt.bar(index, df_opioids['FPOO-Rate'], bar_width, alpha = 0.3, color = 'b', label = 'Opioids')
Heroine = plt.bar(index, df_opioids['FFHO-Rate'], bar_width, alpha = 0.3, color = 'g', label = 'Herione')

plt.xlabel('County Name')
plt.ylabel('Overdoes-Rate per 100,000 County Residents')
plt.xticks(index + bar_width , (df_opioids['CountyName']))
plt.xticks(rotation = 90) #text rotation
ax.xaxis.label.set_size(20)
ax.yaxis.label.set_size(20)
plt.legend()

plt.tight_layout()
plt.show()
```

In order to answer this question, I needed to visualize the data by using a bar chart to see overdoes rate between Opioids and heroine overdoes by county. The data on this bar chart show no correlation between Opioids and Heroine in most counties in Virginia

## How have vehicle sales in the United States varied over time? 

```python
import matplotlib.ticker as ticker
tick_spacing = 5

"""Format of the time series method"""
fig, ax = plt.subplots()
fig.suptitle(' United States Vehicle sales', fontsize = 24) #Graph Title
fig.set_size_inches(24,6)

ax.plot(df_NSA['DATE'],df_NSA['TOTALNSA'], c = 'r' ) #alpha controls dot transparency: 1 = solid , 0 =no mark, b stand for blue
ax.xaxis.set_label_text('Date') # x axis caption
ax.yaxis.set_label_text('Total sales (in Thousand Units) ')
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.xaxis.label.set_size(20)
ax.yaxis.label.set_size(20)

plt.xticks(fontsize = 10, rotation = 90) #text rotation
plt.show()
```

The line graph is best for this exmaple so the viewer can easily read the time series trends
